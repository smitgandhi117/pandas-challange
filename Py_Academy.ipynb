```python
#imports
import pandas as pd
import numpy as np

#read csv files
school_data_to_load = "/Users/Smit/desktop/smit_repo/class_repo/04-Pandas/Instructions/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "/Users/Smit/desktop/smit_repo/class_repo/04-Pandas/Instructions/PyCitySchools/Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#combine the two datasets into a single dataset
school_data_complete = pd.merge(student_data, school_data, on="school_name")

#display the data
display(school_data_complete.head())
```

```python
#Disctrict Summary
#total number of schools
totalSchools = len(school_data_complete.school_name.unique())
print(totalSchools)

#total number of students
totalStudents = len(school_data_complete['Student ID'])
print(totalStudents)

#total budget
totalBudget = np.sum(school_data_complete.budget.unique())
print(totalBudget)

#average math score
avgMathScore = np.mean(school_data_complete.math_score)
print(avgMathScore)

#average reading score
avgReadingScore = np.mean(school_data_complete.reading_score)
print(avgReadingScore)

#overall passing rate; average of math and reading scores
passingRate = (avgMathScore + avgReadingScore)/2
print(passingRate)

#percentage of students with passing math score; 70 or greater
passingMath_rate = (len(school_data_complete.math_score[school_data_complete.math_score >= 70]) / totalStudents) * 100
print(passingMath_rate)

#percentage of students with passing reading score; 70 or greater
passingReading_rate = (len(school_data_complete.reading_score[school_data_complete.reading_score >= 70]) / totalStudents) * 100
print(passingReading_rate)

#creating disctrict dataframe
data_district = [[totalSchools, totalStudents, totalBudget, avgMathScore, avgReadingScore, passingMath_rate, passingReading_rate, passingRate]]
district_summary = pd.DataFrame(data_district, columns = ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing Rate'])

district_summary['Total Students'] = district_summary.apply(lambda x: "{:,}".format(int(x['Total Students'])), axis=1)
district_summary['Total Budget'] = district_summary.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
display(district_summary)
```

```python
#combine the two datasets
data_bySchool = school_data_complete.groupby(["school_name"])
data_bySchool.count()
```

```python
#School Summary

#Create an overview table that summarizes key metrics about each school, using the following:
#school name
schoolName = data_bySchool['school_name']

#school type
schoolType = school_data.set_index('school_name')['type']

#total school students
schoolTotalStudents = data_bySchool['size'].count()

#total school budget
schoolBudget = school_data.set_index('school_name')['budget']

#per student budget
perStudentBudget = schoolBudget / schoolTotalStudents

#average math score 
schoolAvgMathScore = data_bySchool["math_score"].mean()

#average reading score
schoolAvgReadingScore = data_bySchool["reading_score"].mean()

#percentage passing math
schoolPassingMathRate = (school_data_complete[school_data_complete.math_score >= 70].groupby('school_name')['Student ID'].count() / schoolTotalStudents) * 100

#percentage passing reading
schoolPassingReadingRate = (school_data_complete[school_data_complete.reading_score >= 70].groupby('school_name')['Student ID'].count() / schoolTotalStudents) * 100

#overall passing rate
schoolPassingRate = (schoolAvgMathScore + schoolAvgReadingScore) / 2

#create a dataframe that displays schools summary
school_summary = pd.DataFrame({"School Type": schoolType, 
                               "Total Students": schoolTotalStudents,
                               "Total School Budget": schoolBudget,
                               "Budget Per Student": perStudentBudget,
                               "Average Math Score": schoolAvgMathScore,
                               "Average Reading Score": schoolAvgReadingScore,
                               "% Math Passing": schoolPassingMathRate,
                               "% Reading Passing": schoolPassingReadingRate,
                               "Overall Passing Rate": schoolPassingRate})
print("School Summary: ")
display(school_summary)


```

```python
#Top performing schools by passing rate
print("Top Performing Schools Overall Passing Rate:")
school_summary.sort_values(by='Overall Passing Rate', ascending=False).head()
```

```python
#Top performing schools by passing rate
print("Bottom Performing Schools Overall Passing Rate:")
school_summary.sort_values(by='Overall Passing Rate').head()
```

```python
#Math Scores by Grade
ninth = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth = school_data_complete[(school_data_complete["grade"] == "12th")]

freshman = ninth.groupby("school_name")["math_score"].mean()
sophomore = tenth.groupby("school_name")["math_score"].mean()
junior = eleventh.groupby("school_name")["math_score"].mean()
senior = twelfth.groupby("school_name")["math_score"].mean()

math_by_grade = pd.DataFrame({"9th": freshman, 
                               "10th": sophomore,
                               "11th": junior,
                               "12th": senior})
print("Math Scores by Grade: ")
display(math_by_grade)
```

```python
#Reading Scores by Grade
ninth = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth = school_data_complete[(school_data_complete["grade"] == "12th")]

freshman = ninth.groupby("school_name")["reading_score"].mean()
sophomore = tenth.groupby("school_name")["reading_score"].mean()
junior = eleventh.groupby("school_name")["reading_score"].mean()
senior = twelfth.groupby("school_name")["reading_score"].mean()

reading_by_grade = pd.DataFrame({"9th": freshman, 
                               "10th": sophomore,
                               "11th": junior,
                               "12th": senior})
print("Reading Scores by Grade: ")
display(reading_by_grade)
```

```python
#Scores by School Spending
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_summary["School Spending"] = pd.cut(perStudentBudget, spending_bins, labels=group_names)

avgMathScore = school_summary.groupby(["School Spending"]).mean()['Average Math Score']
avgReadingScore = school_summary.groupby(["School Spending"]).mean()['Average Reading Score']
percPassingMath =  school_summary.groupby(["School Spending"]).mean()['% Math Passing']
percPassingReading =  school_summary.groupby(["School Spending"]).mean()['% Reading Passing']
passingRate = (percPassingMath + percPassingReading)/2

scores_by_spending = pd.DataFrame({"Average Math Score": avgMathScore, 
                               "Average Reading Score": avgReadingScore,
                               "% Math Passing": percPassingMath,
                               "% Reading Passing": percPassingReading,
                                   "% Overall Passing Rate": passingRate})

print("Scores by School Spending Per Student: ")
display(scores_by_spending)
```

```python
#Scores by School Size
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)

avgMathScore = school_summary.groupby(["School Size"]).mean()['Average Math Score']
avgReadingScore = school_summary.groupby(["School Size"]).mean()['Average Reading Score']
percPassingMath =  school_summary.groupby(["School Size"]).mean()['% Math Passing']
percPassingReading =  school_summary.groupby(["School Size"]).mean()['% Reading Passing']
passingRate = (percPassingMath + percPassingReading)/2

scores_by_size = pd.DataFrame({"Average Math Score": avgMathScore, 
                               "Average Reading Score": avgReadingScore,
                               "% Math Passing": percPassingMath,
                               "% Reading Passing": percPassingReading,
                                   "% Overall Passing Rate": passingRate})

print("Scores by School Size: ")
display(scores_by_size)
```
